By my understanding:
1. CRIM: per capita crime rate by town
Tells us about the safety of a town. And higher values indicate higher crime rate which could negatively impact housing prices.

2. ZN : proportion of residential land zoned for lots over 25,000 sq.ft.
Tells us about the availability of plots of land zoned for residential purpose. High values suggest more spacious or suborban places.

3. INDUS: proportion of non-retail business acres per town
Non-retail business, tells us that this is mostly an industrial area, and higher values could be associated with lower housing prices.

4. CHAS: charles river dummy variable
Tells us wehter a house is close to the charles river. Houses by the river attract more money because of its desirability.

5. NOX: nitric oxides concentration
Tells us about the air quality, with higher values indicating a poluted air which could negatively impact house prices.

6. RM: average number of rooms per dwelling
Tells us the number of rooms in the house. Generally, houses with more rooms are more expensive than with less.

7. Age: proportion of owner-occupied units built prior to 1940
Tells us about the age of the house. Older houses may be less expensive due to wear/tear but may also have historical value

8. DIS: weighted distances to five Boston employment centres
This feature measures how far the houses are from employment centers. Greater distances could lead to lower housing prices due to less accessibility.

9. RAD: index of accessibility to radial highways
This index measures accessibility to main highways, potentially increasing housing prices due to ease of travel.

10. TAX: full-value property-tax rate per $10,000
Higher tax rates might be associated with more affluent areas but could also decrease housing affordability and desirability because even the rich dont like to pay tax - sometimes lol.

11. PTRATIO: pupil-teacher ratio by town
Lower pupil-teacher ratios are often indicative of better school quality, which can positively influence housing prices.

12. B (calculated as B=1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town:
This is a measure used historically to assess the proportion of the black population in the area. We can use this to explore the impact of demography on housing prices.

13. LSTAT: % lower status of the population
This shows like the socio-economic status of the population. Lower values would likely indicate more expensive homes.
Simply because poor people cant afford expensive homes.

14. MEDV: Median value of owner-occupied homes in $1000's
This is my target variable showing the median value of homes in a particular area.

#### EDA Narrative 1:
Narrative 1: Geographical impact and house prices. Does proximity to the Charles River (CHAS) and accessibility to 
radial highways (RAD) affect house prices?

```python
# CHAS and MEDV
# lets see the mean and median house prices for areas with and without proximity to the CHAS river.

chas_statistics = df.groupby('CHAS')['MEDV'].agg(['mean', 'median'])
print('Mean and Median of houses with proximity to Charles River\n')
print(chas_statistics)

# box plot for CHAS vs MEDV
plt.figure(figsize= (8, 6), dpi = 100)
sns.boxplot(x = 'CHAS', y = 'MEDV', data= df)
plt.title('House Prices in relation to proximity to Charles River')
plt.xlabel('0: Not Close to Charles, 1 : Close to Charles')
plt.ylabel('Mean House Price per $1k')

# statistical test to confirm
from scipy import stats
ttest = stats.ttest_ind(df[df['CHAS'] == 1]['MEDV'], df[df['CHAS'] == 0]['MEDV'])

print(f'\nT-test for CHAS: statistic= {ttest.statistic}, p_value = {ttest.pvalue}')

'''
Jacobs Interpretation: Generally, houses that are close to the Charles river are more expensive than the farther. Several 
reasons include desirability, better view and scenery, cooler temperatures during the Boston summer, among others. With a 
low p-value, the difference in the mean of the two groups is likely real and not due to random chance.
'''
```

#### EDA Narrative 3:
Narrative 3: Urban Vs. Sub-urban. How does living urban industralized neighbourhood or suburbs affect house prices?

```python
# ZN & INDUS vs. MEDV
print('Description of ZN and INDUS')
df[['ZN', 'INDUS']].describe()
```

```python
# If an area has above the mean of ZN; then we can assume that that area is suburban
# If an area has above the mean of INDUS; then we can assume that that area is Industralized.

df_n3 = df.copy() # create a copy of our dataframe into a new dataframe called df_narrative3. I do this to avoid data corruption


def area_evaluator(data): #fcn to engineer new feature of suburban or urban
    if data['ZN'] > 11:
        return 'Suburban'
    elif data['INDUS'] > 11:
        return 'Urban'
    else:
        return 'Other'
    
# create a new dataframe called AreaType
df_n3['AreaType'] = df_n3.apply(area_evaluator, axis= 1)

# groupby
area_stats = df_n3.groupby('AreaType')['MEDV'].agg(['mean', 'median'])
print('Mean and Median of House Prices Based on their Area Type')
print(area_stats)

# swarmplot
plt.figure(figsize= (10, 8), dpi= 100)
sns.violinplot(x = 'AreaType', y = 'MEDV', data= df_n3)
plt.title('House prices by Area Type')


'''
Jacobs Note: the mean house prices in the suburbs is 36% higher than that of urban industralized areas.
'''
```

```python
# Identifying and handling outliers
def handle_outliers(df):
    for col in df.columns[:]:
        if pd.api.types.is_numeric_dtype(df[col]):
            Q1 = df[col].quantile(0.25)
            Q3 = df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
            print(f'{col}: {len(outliers)} outliers')
        
handle_outliers(df)
'''
Jacobs Note: Outliers exists in the dataset. In this particular case, I want to work based on the assumption these outliers
are not a result of handling/imputation error during data capture. And the number of outliers that exists in each column 
is relatively okay. With these assumptions, I take the decision to decide that these outliers true anomalies that are 
representative of the real world, and I will do nothing. In other words, I will handle the outliers by not handling the 
outliers.
'''
```

## Phase 3: Feature Engineering

Due to the nature of my choice dataset, a few features can be engineered from existing features. Such as binning the age of the houses. We will categorize houses as old, moderately old, and new.

```python
df['AGE'].describe()
```

```python
# create a new dataframe called df_fe

df_fe = df.copy()

def age_bin(row): #fcn to categorize the house age
    if row['AGE'] < 15:
        return 'new'
    elif row['AGE'] < 65:
        return 'moderate'
    else:
        return 'old'
    
# create a new dataframe called AreaType
df_fe['AgeBin'] = df_fe.apply(age_bin, axis= 1)
df_fe.head()

'''
Jacobs Note: Although AgeBin doesn't necessarily help our regression analysis, it gives us a quick view on the how the 
age of the houses relate to their price.
'''
```

```python
print('Describe of out feature engineering data\n')
print(df_fe.describe())

'''
Lets standardize the columns with the large variance, namely, ZN and B.
'''

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # initialize a Standard Scaler object

df_fe['stdCRIM'] = scaler.fit_transform(df[['CRIM']])
df_fe['stdZN'] = scaler.fit_transform(df[['ZN']])
df_fe['stdINDUS'] = scaler.fit_transform(df[['INDUS']])
df_fe['stdCHAS'] = scaler.fit_transform(df[['CHAS']])
df_fe['stdNOX'] = scaler.fit_transform(df[['NOX']])
df_fe['stdRM'] = scaler.fit_transform(df[['RM']])
df_fe['stdAGE'] = scaler.fit_transform(df[['AGE']])
df_fe['stdDIS'] = scaler.fit_transform(df[['DIS']])
df_fe['stdRAD'] = scaler.fit_transform(df[['RAD']])
df_fe['stdTAX'] = scaler.fit_transform(df[['TAX']])
df_fe['stdPTRATIO'] = scaler.fit_transform(df[['PTRATIO']])
df_fe['stdB'] = scaler.fit_transform(df[['B']])
df_fe['stdLSTAT'] = scaler.fit_transform(df[['LSTAT']])
df_fe['stdMEDV'] = scaler.fit_transform(df[['MEDV']])

print('\nStandardized dataset')
print(df_fe.head(3))
```

## Phase 4: Model Training and Evaluation

We will first train a linear regression model first as it is quite intuitive that we are trying to predict a continuous variable which is house prices.
<br>
We will also attempt to see the effect of standardizing the columns on our linear model.

##### Linear Model

```python
# import training test split
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df.drop(columns='MEDV')
y = df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state= 42)
```

```python
# train and fit a linear regression model on the training set
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression() #instatiate the object
lm.fit(X_train, y_train) # fit the train data to the model
```

```python
# Coefficients
print('Coefficient of linear model: \n', lm.coef_)
```

```python
# Predicting the test data
predictions = lm.predict(X_test)
```

```python
# Create a scatterplot of the real test values vs. the predicted values
plt.scatter(y_test, predictions)
plt.xlabel('y test')
plt.ylabel('predicted y')
```

```python
# evaluating the model
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
# coeffecients
coeffecients = pd.DataFrame(lm.coef_, X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients

'''
Jacobs Interpretation: For our linear regression model,
Holding all other features fixed, a 1 unit increase in CRIM is associated with a -0.133 decrease in target variable.
Holding all other features fixed, a 1 unit increase in ZN is associated with a 0.035 increase in target variable.
Holding all other features fixed, a 1 unit increase in INDUS is associated with a 0.049 increase in target variable.
Holding all other features fixed, a 1 unit increase in CHAS is associated with a 3.11 increase in target variable.
Holding all other features fixed, a 1 unit increase in NOX is associated with a -15.417 decrease in target variable.

..ETC...

'''
```

##### Linear Model with all columns standardized

```python
# Lets grab our feature engineered dataset; the dataset with the standardized columns
df_fe.columns
```

```python
# Select our X and y
X_fe = df_fe.drop(columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV', 'AgeBin', 'stdMEDV']) #drop columns we dont need
y_fe = df_fe['stdMEDV']
```

```python
Xfe_train, Xfe_test, yfe_train, yfe_test = train_test_split(X_fe, y_fe, test_size=0.3, random_state= 42)
```

```python
# train and fit a linear regression model on the training set
from sklearn.linear_model import LinearRegression
```

```python
lm_fe = LinearRegression() #instatiate the object
lm_fe.fit(Xfe_train, yfe_train) # fit the train data to the model
fe_coeffecients = pd.DataFrame(lm_fe.coef_, X_fe.columns)
fe_coeffecients.columns = ['Coeff of standardized variable']
print(fe_coeffecients)


# predictions
fe_predictions = lm_fe.predict(Xfe_test)

# evaluating the model
from sklearn import metrics

print('std Linear MAE:', metrics.mean_absolute_error(yfe_test, fe_predictions))
print('std Linear MSE:', metrics.mean_squared_error(yfe_test, fe_predictions))
print('std Linear RMSE:', np.sqrt(metrics.mean_squared_error(yfe_test, fe_predictions)))

'''
Jacobs Note: Standardization: The significantly lower error metrics (MAE, MSE, and RMSE) after standardization indicate t
hat the model's predictions are much closer to the actual values compared to the predictions made using the original, 
non-standardized data. This suggests that standardization has helped improve the model's performance, likely by ensuring 
that each feature contributes proportionately to the prediction outcome.
'''
```

##### Random Forest

A random forest regressor is the choice for a second model because of its abilitiy to handle complexities like non linear relationships, and it is less affected by outliers - which exists in this dataset

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
```

```python
rfr_pred = rfr.predict(X_test)

# evaluating the model
from sklearn import metrics

print('Random Forest MAE:', metrics.mean_absolute_error(y_test, rfr_pred))
print('Random Forest MSE:', metrics.mean_squared_error(y_test, rfr_pred))
print('Random Forest RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rfr_pred)))

# Get feature importances
feature_importances = rfr.feature_importances_

# Create a DataFrame for better visualization
feature_importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(feature_importances_df)

'''
Jacobs Note: The random forest regressor have outperformed our linear regression model in prediciting the house price.
As we have lower values of MAE, MSE, RMSE on our non-standardized dataset.
'''
```

```python
# random forest regressor on standardized data (df_fe)
rfr_fe = RandomForestRegressor(random_state=42)
rfr_fe.fit(Xfe_train, yfe_train) #fit model on standardized data

rfr_fe_pred = rfr_fe.predict(Xfe_test) #predict

#evaluate the model
print('Std Random Forest MAE:', metrics.mean_absolute_error(yfe_test, rfr_fe_pred))
print('Std Random Forest MSE:', metrics.mean_squared_error(yfe_test, rfr_fe_pred))
print('Std Random Forest RMSE:', np.sqrt(metrics.mean_squared_error(yfe_test, rfr_fe_pred)))

# Get feature importances
feature_importances = rfr_fe.feature_importances_

# Create a DataFrame for better visualization
feature_importances_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print(feature_importances_df)

'''
Jacobs Note: The random forest regressor still outperforms our linear regression model in the standardized dataset.
'''
```
